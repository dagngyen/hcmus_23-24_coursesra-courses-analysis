## Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from itertools import product
import numpy as np
import seaborn as sns
import time

import warnings
warnings.filterwarnings('ignore')
```

## Evaluate model on the best hyperparameters

+ Read data from csv

```python
df = pd.read_csv('evaluation.csv', header=None, names=['time', 'mse', 'r2-score'])
df.index = ['Linear with SGD', 'Neural Network', 'k-NN', 'SVM']
df
```

+ Visualize the evaluation of 4 models

```python
ax = plt.bar(df.index, df['time'])
plt.ylabel('Time')
plt.title('Runtime of models')
plt.bar_label(ax);
```

+ We see that among the 4 models, the **k-NN** model has the fastest running time and the **Neural Network** model has a fairly large running time.
+ Furthermore, the **Linear regression with SGD** and **SVM** has relatively good running time.
+ This proves that the **Neural Network** is not optimal for small datasets (~1000 objects), the remaining 3 models are more suitable for training with this dataset.

```python
ax = plt.bar(df.index, df['mse'])
plt.ylabel('Time')
plt.title('Mean Squared Error of models')
plt.bar_label(ax);
```

+ Once again, MSE of the **Neural Network** model is the highest, **k-NN** is at a relative level. The two **Linear regression models with SGD** and **SVM** models nearly the same MSE at a fairly low level.

```python
ax = plt.bar(df.index, df['r2-score'])
plt.ylabel('Time')
plt.title('R-square score of models')
plt.bar_label(ax);
```

+ Most models show an r-square score approaching 0.4, however **Neural Network** is only at 0.2. Once again confirming that the **Neural Network** is ineffective.
+ Model **Linear regression with SGD** gives the highest r-square score, inferring that this is the appropriate model for the current data set.

## Conclusion

+ Training process:
     + **k Nearest Neighbor** gives very small training time and is the fastest.
     + **Linear regression with SGD** and **Support Vector Machine** has stable training time.
     + **Neural network** takes longer to train than the other 3 models.
+ Testing evaluation:
     + **Linear regression with SGD** and **Support Vector Machine** has the smallest MSE and relatively high r-square.
     + Although **k Nearest Neighbor** has a fast training speed, the MSE is larger and the r-square is not too impressive.
     + **Neural network** has the largest MSE and the smallest r-square. This can mean that the **Neural network** is not suitable for the data set.
+ In summary:
     + For a small dataset, **Neural Network** is very inefficient and becomes very likely to overfit, we can apply some 'early-stopping' methods to improve this.
     + **Linear Regression with SGD** works well, but might lose accuracy when the dataset becomes larger and more complex.
     + **k Nearest Neighbor** does not seem to take time to learn from the training set. However, with large data, the performance of k-NN will be affected because it must save all the data.
     + **SVM** is relatively optimal because it provides impressive results whether considering MSE or r-square. When faced with the case where the number of attributes of the data set is much larger than the number of data, **SVM** gives quite poor results.
