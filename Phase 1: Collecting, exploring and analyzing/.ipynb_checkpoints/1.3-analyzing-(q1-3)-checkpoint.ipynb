## Import necessary libraries

```python
import os
import sys

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Read data from csv

```python
df = pd.read_csv('../Data/processed_data.csv', sep=',',engine='python', encoding='utf-8')
df.head(10)
```

## Question 1: Find out the fond of people on the duration of courses. Will the duration affect the choice?

- **Answer this question**: In order to make more suitable courses for learner, we need to find the fond of people in duration of courses - one aspect of a courses, so that we can have more attractive courses for people to enroll
- **To Answer this question**: Use `'enrollment'`, `'duration'` and `'general'` columns to find the answer

```python
duration_df = df[['enrollment', 'duration', 'general']]
duration_df['rating_category'] = duration_df.apply(lambda row: 'Short' if row['duration'] < 32 else 'Medium' if 32 < row['duration'] < 65 else 'Long' if row['duration'] > 65 else None, axis=1)
duration_df
```

```python
most_common_value = duration_df['general'].mode().iloc[0]
only_field_df = duration_df[duration_df['general'] == most_common_value]
only_field_df
```

```python
en_du_df = only_field_df.groupby(['rating_category'])['enrollment'].agg(['mean'])
en_du_df.columns = ['average_enrollment']
en_du_df
```

```python
explode = (0.1, 0.1, 0.1)  
fig, ax = plt.subplots()
ax.pie(en_du_df['average_enrollment'], explode=explode, labels=en_du_df.index, autopct='%1.1f%%',shadow={'ox': -0.05, 'edgecolor': 'none', 'shade': 1}, startangle=100)
plt.title('Enrollment Distribution by Duration in Coursera at most-courses Field')
plt.show()
```

**Conclusion**
- In analyzing phase before plotting above chart, we find that `Business` field courses is more attractive than other field
- As the chart above, We can easily see that people prefer on `medium` and `short` duration courses than `long` duration courses
- With the information, We can create suitable and attractive courses for people to join in. It helps shaping courses

## Question 2: Evaluate the courses offered by university if the recommend is good or not

- **Answer this question**: help us know more infomation about the `quality of courses` that `offered by university/institude`. Not all recommandation are good, we need to check more information about that courses instead of just only listening to the offer
- **To answer this question**: use `'rating'` columns to find the mean of rating to compare, `'offered by'`, `'rating'` to find the mean of recommendation's ration from university/institude

```python
avg_rating = df['rating'].mean()
school_offer = df.groupby(['offered by'])['rating'].agg(['mean'])
school_offer.columns = ['average_rating']
school_offer = school_offer.sort_values(by='average_rating')
school_offer['rating_category'] = ['Lower' if x < avg_rating else 'Equal or Higher' for x in school_offer['average_rating']]
school_offer['comparing_rating'] = school_offer['average_rating'] - avg_rating
school_offer
```

```python
pie_offered = school_offer['rating_category'].value_counts()
fig, ax1 = plt.subplots()
ax1.pie(pie_offered.values, labels=pie_offered.index, autopct='%1.1f%%',shadow={'ox': -0.05, 'edgecolor': 'none', 'shade': 1}, startangle=100)
plt.title('Percentage about Status of Average Rating of school when compare to average rating of all courses')
plt.show()

fig, ax2 = plt.subplots()
ax2.bar(school_offer.index, school_offer['comparing_rating'], label='Rating from DataFrame')
ax2.axhline(y=0, color='red', linestyle='-', label='average_rating')
ax2.set_xlabel('University/Institute')
ax2.set_ylabel('Rating')
ax2.set_title('Rating of Courses offered when compare with the average rating of all courses')
ax2.legend()
ax2.set_xticks([])
```

**Conclusion**
- Base on the Pie chart, We can see that more than 50% recommendation by university/institude has `higher or equal rating` than `avarage rating` of all courses.
- With the second chart, it shows us that not all recommandation are good. Some recommendations have a `lower rating` than the `average rating` of all courses. Besides that, there are still many good recommandation, and it help to prove the conclusion above is correct when part show higher rating has more space than lower rating

## Question 3: Check the general level of learner in these years

- **Answer this question**: Help you to know the average level in social these day. It can show you if people like learn new things than go deeper in an aspect or not.
- **To answer this question**: use `'level'` and `'enrollment'` to answer this question

```python
general_level = df.groupby(['level'])['enrollment'].agg(['sum'])
general_level.columns = ['total_enrollment']
general_level
```

```python
explode = (0.1, 0.1, 0.1, 0.1)  
fig, ax = plt.subplots()
ax.pie(general_level['total_enrollment'], explode=explode, labels=general_level.index, autopct='%1.1f%%',shadow={'ox': -0.05, 'edgecolor': 'none', 'shade': 1}, startangle=100)
plt.title('Enrollment Distribution by Course Level')
plt.show()
```

**Conclusion**
- From the pie chart above, We can see more than 50% people attempt on `Beginner` Courses while `Advanced` level has least enrollment
- Answer this question can help showing the level of people in common or find out people tend to master one aspect or learn more knowlegde beside their strongest aspect. This also show the preference of people on the level of courses. From that we can shape the courses to attract more enrollments
